## Visualization for univariate analysis

```python
#This function takes the numerical column as the input and returns the boxplots and histograms for the variable
#to check spread, central tendency , dispersion and outliers.
def histogram_boxplot(data, figsize=(15,10)):
    f2, (ax_box, ax_hist) = plt.subplots(nrows = 2, # Number of rows of the subplot grid= 2
                                           sharex = True, # x-axis will be shared among all subplots
                                           gridspec_kw = {"height_ratios": (.25, .75)}, 
                                           figsize = figsize 
                                           ) # creating the 2 subplots
   
    sns.boxplot(data, ax=ax_box, showmeans=True, color='blue') # For boxplot 
    sns.distplot(data, kde=False, color='Blue', ax=ax_hist)# For histogram
    ax_hist.axvline(data.mean(), color='y', linestyle='--') # Add mean to the histogram
    ax_hist.axvline(data.median(), color='r', linestyle='-') # Add median to the histogram
    ax_hist.axvline(data.mode()[0],color='g',linestyle='-') # Add mode to the histogram
    ax_hist.legend({'Mean':data.mean(),'Median':data.median(),'Mode':data.mode()})
```

```python
# Function to create barplots that indicate percentage for each category.
def bar_perc(plot, feature):
    '''
    plot
    feature: 1-d categorical feature array
    '''
    total = len(feature) # length of the column
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total) # percentage of each class of the category
        x = p.get_x() + p.get_width() / 2 - 0.05 # width of the plot
        y = p.get_y() + p.get_height()           # height of the plot
        ax.annotate(percentage, (x, y), size = 12) # annotate the percantage
```

```python
plt.figure(figsize=(10,7))
ax = sns.countplot(fitness['Product'])
plt.xlabel('Product')
plt.ylabel('Count')
bar_perc(ax,fitness['Product'])
```

- Model TM195 is the most purchased model, while model TM498 is purchased more than model TM798.
- 44.4% customers bought model TM195, 33.3% customers bought model TM498 and 22.2% customers bought model TM798.

```python
plt.figure(figsize=(10,7))
ax = sns.countplot(fitness['Gender'])
plt.xlabel('Gender')
plt.ylabel('Count')
bar_perc(ax,fitness['Gender'])
```

- There are 57.8% male customers that purchased treadmills, while 42.2% female customers that purchased treadmills

```python
plt.figure(figsize=(10,7))
ax = sns.countplot(fitness['MaritalStatus'])
plt.xlabel('MaritalStatus')
plt.ylabel('Count')
bar_perc(ax,fitness['MaritalStatus'])
```

- There are 59.4% partnered customers that purchased treadmills, while 40.6% single customers that purchased treadmills.
